## Import necessary Library

```python
import pandas as pd
import os
```

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")
```

```python
files =[file for file in os.listdir("./Sales_Data/")]
```

```python
all_month_data = pd.DataFrame()
for file in files:
    df = pd.read_csv(f"./Sales_Data/{file}")
    all_month_data = pd.concat([all_month_data, df])

all_month_data
```

```python
from pathlib import Path
dir_data = Path("./Sales_Data/")
all_month_data = pd.concat([pd.read_csv(f) for f in dir_data.iterdir()])
```

```python
nan_data = all_month_data[all_month_data.isnull().any(axis = 1)]
nan_data
all_month_data = all_month_data.dropna(how = "all")
all_month_data
```

```python
all_month_data = all_month_data[all_month_data["Order Date"].str[0:2] != "Or"]
```

```python
all_month_data["Month"] = all_month_data["Order Date"].str[0:2].astype("int32")
```

```python
all_month_data.head()
```

```python
all_month_data.info()
```

```python
all_month_data["Quantity Ordered"] = pd.to_numeric(all_month_data["Quantity Ordered"])
all_month_data["Price Each"] = pd.to_numeric(all_month_data["Price Each"])
```

```python
all_month_data["Sales"] = all_month_data["Quantity Ordered"]*all_month_data["Price Each"]
```
