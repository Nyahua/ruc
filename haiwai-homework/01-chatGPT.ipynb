```python
df = pd.read_csv('city.csv', index_col=0)

def text_clean(text: str):
    text = text.replace("\u3000", ' ')
    text = re.sub(' +', ' ', text)
    text = re.sub('\n+', '\n', text)
    return text
for city in df.index:
    text = df.loc[city, 'text']
    context = "## CONTEXT: \n\n" + text_clean(text)[:2000]
    prompt = (
        f"## According to the following CONTEXT, answer the question bellow:"
        + f"{context}"
        + f"## Question:\n# {city}第二产业增加值为: <answer>"
    )
    answer = openai.Completion.create(
        prompt=prompt,
        temperature=0.0,
        max_tokens=300,
        top_p = TOP_P,
        engine='text-davinci-003'
    )["choices"][0]["text"].strip(" \n")
    print(city, answer)
    try:
        answer = re.sub(r'2020' , '', answer)
        answer_num = float(re.sub(r'[^0-9.]' , '', answer))
    except:
        answer_num = np.nan
    df.loc[city, ['GDP', 'GDPnum']] = answer, answer_num
    sleep(1)
    
df.to_csv('city-gdp.csv')
```
