```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
from statsmodels.graphics.mosaicplot import mosaic

mpl.rcParams['font.family'] = ['Microsoft YaHei']
```

```python
data = pd.read_excel("data/china-C06-21.xlsx", sheet_name="data", skiprows=6)
data = data[data["地区"].isin(["北京", "天津", "上海"])]
data = data.set_index("地区")
data
```

```python
gender = ['male', 'male', 'male', 'female', 'female', 'female']
pet = ['cat', 'dog', 'dog', 'cat', 'dog', 'cat']
data = pd.DataFrame({'gender': gender, 'pet': pet})
mosaic(data, ['pet', 'gender'], title='DataFrame as Source')
plt.show()
data
```

```python
rand = np.random.random
from itertools import product
tuples = list(product(['bar', 'baz', 'foo', 'qux'], ['one', 'two']))
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
data = pd.Series(rand(8), index=index)
mosaic(data, title='hierarchical index series')
plt.show()
data
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
data = pd.read_excel("data/china-C06-21.xlsx", sheet_name="data", skiprows=6)
data = data[data["地区"].isin(["北京", "天津", "上海"])]
data = data.set_index("地区")
data = data.stack()
mosaic(data, ax=ax, labelizer=lambda k: '', horizontal=False, label_rotation=90)
plt.show()
```
