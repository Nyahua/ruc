```python
# 参与人数 去掉一个汉字“亿"
df = pd.read_excel("./数据集.xlsx", skiprows=1, nrows=2, usecols="B:H", index_col=0)
df.loc["参与人数", :] = df.loc["参与人数", :].str[:-1].str.strip().astype(float)
df
```

```python
plt.bar(df.columns, df.loc["参与人数", :])
```
