## import necessary libararies

```python
import numpy as np
import pandas as pd
from pathlib import Path
```

## combine all CSV files to one dataframe

```python
dir_data = Path("./Sales_Data/")
df = pd.concat([pd.read_csv(f, index_col=0) for f in dir_data.iterdir()])
print(df.shape)
df.head()
```

### Data Cleaning

```python
#finding rows contains any missing values
# df[df.isna().any(axis=1)]
# drop all nan rows
df = df.dropna(how="all")

# delete rows contains column names
df = df[~df['Order Date'].str.contains("Or")]

# convert "Quantity Ordered" to int
df["Order_Qty"] = df["Quantity Ordered"].astype(int)
```

## convert string to datatime

```python
df["Order_Date"] = pd.to_datetime(df['Order Date'])
```

```python
df.groupby(df.Order_Date.dt.month.rename("month"))["Order_Qty"].sum()
```
